# Data Visualization Practice

## Call Libs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset

### 1. Monthly Sales (for Line Chart)

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1700, 1600, 1800, 2000]
```

### 2. Product Revenue (for Bar Chart)

```python
products = ["Laptops", "Mobiles", "Tablets", "Accessories"]
revenue = [35000, 50000, 15000, 10000]
```

### 3. Market Share (for Pie Chart)

```python
languages = ["Python", "Java", "C++", "JavaScript"]
popularity = [40, 25, 20, 15]
```

## Tasks

### 1. Line Chart

```python
plt.plot(months, sales, color = 'purple', marker = 'o', label = 'Sales')

plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
```

### 2. Bar Chart

```python
plt.bar(products, revenue, color = 'pink', label = 'Revenue' )

plt.title('Revenue by Product')
plt.xlabel('Products')
plt.ylabel('Revenue')
plt.legend()
```

### 3. Pie Chart

```python
plt.pie(popularity, labels = languages, autopct = '%1.1f%%', startangle = 90)

plt.title('Programming Language Market Share')
```
