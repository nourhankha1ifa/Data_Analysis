```python
import numpy as np

# 1
arr = np.random.randint(10, 101, size=20)
print(arr)
```

```python
# 2
print("First 5 elements:", arr[:5])
print("Last 3 elements:", arr[-3:])
print("Elements at even indices:", arr[::2])
```

```python
# 3
arr2 = np.random.randint(1, 11, size=20)
print("Second array:", arr2)

print("\nAddition:")
print(arr + arr2)

print("\nSubtraction:")
print(arr - arr2)

print("\nMultiplication:")
print(arr * arr2)

print("\nDivision:")
print(np.round(arr / arr2, 2))
```

```python
# 4
matrix = arr.reshape(4, 5)
print("\nReshaped 4x5 matrix:")
print(matrix)

print("\nSecond row:", matrix[1])

print("\nLast column:", matrix[:, -1])

print("\nSub-matrix:")
print(matrix[:2, :3])
```
