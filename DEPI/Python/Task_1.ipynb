# DATA TYPES AND VARIABLES

```python
# A) Convert Fahrenheit to Celsius
fah = float(input("Enter temperature in Fahrenheit: "))
cel = (fah - 32) * 5 / 9
print(f"Celsius = {cel:.2f}")
```

```python
# B) Area of a Circle
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius ** 2
print(f"Area of circle = {area:.2f}")
```

# OUTPUT / INPUT IN PYTHON

```python
# A) Greeting Message
name = input("Enter your name: ")
print(f"Hello, {name}!")
```

```python
# B) Calculate Birth Year
age = int(input("Enter your age: "))
birth_year = 2026 - age
print(f"You were born in: {birth_year}")
```

```python
# C) Area of a Rectangle
length = float(input("Enter the length of the rectangle: "))
width  = float(input("Enter the width of the rectangle: "))
rect_area = length * width
print(f"Area of rectangle = {rect_area:.2f}")
```

# MATHEMATICAL OPERATIONS

```python
# A) Arithmetic Operations on Two Numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print(f"\nAddition       : {num1} + {num2} = {num1 + num2}")
print(f"Subtraction    : {num1} - {num2} = {num1 - num2}")
print(f"Multiplication : {num1} * {num2} = {num1 * num2}")

if num2 != 0:
    print(f"Division       : {num1} / {num2} = {num1 / num2:.2f}")
else:
    print("Division       : Cannot divide by zero!")
```
