```python
# Write a program to convert Fahrenheit to Celsius.
# Formula: Celsius = (Fahrenheit - 32) × 5/9

# write code here
Fahrenheit = float(input("Enter Fahrenheit degree: "))
Celsius = (Fahrenheit - 32) * 5/9

print("Fahrenheit:", Fahrenheit)
print("Celsius:", Celsius)
```

```python
# Write a program that takes a number as input and calculates both its square and cube.

# write code here
num = float(input("Enter a number: "))
square = num ** 2
cube = num ** 3

print("Number Squared:", square)
print("Number Cubed:", cube)
```

```python
# Write a program that converts kilometers to miles. The program should ask the user for the distance in kilometers.

# write code here
km = float(input("Enter the distance in kilometers: "))
mile = km * 0.621371

print("Distance in miles:", mile)
```

```python
# Write a program to find the larger of two numbers.

# write code here
num_1 = float(input("Enter First Number: "))
num_2 = float(input("Enter Second Number: "))

if num_1 > num_2:
  print("Number 1 is bigger than Number 2")
elif num_1 < num_2:
  print("Number 2 is bigger than Number 3")
else:
  print("Incorrect Input")
```

```python
# Write a program to calculate the total cost of purchasing multiple items. The program should ask for the price and quantity of each item.

# write code here
price = float(input("Enter item price: "))
qty = float(input("Enter item quantity: "))
total_cost = price * qty

print("Total cost:", total_cost)
```

```python
# Write a program to calculate the final grade based on the scores in three subjects. The program should calculate the average score and assign a grade based on it.

# write code here
sub_1 = float(input("Enter First grade: "))
sub_2 = float(input("Enter Second grade: "))
sub_3 = float(input("Enter Third grade: "))

sum = sub_1 + sub_3 + sub_2
avg = sum / 3

print("Final grade:", avg)
```

```python
# Write a program to find the area of a triangle. The program should ask the user for the base and height of the triangle.

# write code here
base = float(input("Enter base: "))
height = float(input("Enter height: "))

area = 0.5 * base * height

print("Area of a triangle:", area)
```
